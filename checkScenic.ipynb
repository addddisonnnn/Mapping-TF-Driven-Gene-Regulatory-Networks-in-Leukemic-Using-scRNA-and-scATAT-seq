```python
# 1. Install requests (one time)
!pip install requests

# 2. Download your leukemia data
import requests
import os

os.makedirs("leukemia_data", exist_ok=True)
url = "https://cf.10xgenomics.com/samples/cell-exp/3.0.0/bmmc_5k_leukemia/bmmc_5k_leukemia_filtered_feature_bc_matrix.h5"
response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})

with open("leukemia_data/bmmc_5k_leukemia.h5", 'wb') as f:
    f.write(response.content)

print(f"✅ Leukemia dataset downloaded! Size: {len(response.content)/1e6:.1f} MB")

# 3. Load it and verify
import scanpy as sc
adata_leukemia = sc.read_10x_h5("leukemia_data/bmmc_5k_leukemia.h5")
print(f"Cells: {adata_leukemia.n_obs}, Genes: {adata_leukemia.n_vars}")
```
